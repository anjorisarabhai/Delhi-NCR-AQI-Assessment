```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("master_aqi_with_hyperlocal_features.csv")
df.head()
```

```python

# --- Preprocessing steps ---

# Parse date column
df['date'] = pd.to_datetime(df['date'], errors='coerce', dayfirst=True)

df.sample(5)
```

```python
# Drop rows with invalid dates
df = df.dropna(subset=['date'])
print("Number of rows after dropping invalid dates:", len(df))
```

```python
df.isnull().sum()
```

## Missing Values Imputation

```python
# Fill missing values for satellite columns using backward fill
# Sort by location and date to ensure proper backward fill
df = df.sort_values(by=['location', 'date'])

# Apply backward fill for satellite columns (NO2, SO2, CO, O3)
satellite_cols = ['NO2_satellite', 'SO2_satellite', 'CO_satellite', 'O3_satellite', 'Aerosol_Index_satellite']
df[satellite_cols] = df.groupby('location')[satellite_cols].bfill()

print("Missing values after backward fill for satellite columns:")
print(df[satellite_cols].isnull().sum())
```

```python
# Impute missing values for hyperlocal features using mean
hyperlocal_cols = [
    'pct_green',
    'pct_industrial',
    'pct_residential',
    'building_density',
    'avg_building_area_m2',
    'median_building_area_m2',
    'building_count',
    'major_road_fraction'
]

# Fill missing values with column mean
for col in hyperlocal_cols:
    if col in df.columns:
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)
        print(f"Filled {col} missing values with mean: {mean_value:.6f}")

print("\nMissing values after mean imputation for hyperlocal features:")
print(df[hyperlocal_cols].isnull().sum())
```

```python
df.isnull().sum()
```

## Temporal Features

```python

# Temporal Features
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek

def get_season(m):
    if m in [12,1,2]:
        return "Winter"
    elif m in [3,4]:
        return "Pre-Monsoon"
    elif m in [5,6,7,8,9]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

df['season'] = df['month'].apply(get_season)

```

```python
df.sample(5)
```

```python

# Target variable: Option A (Ratio)
df['NO2_ratio'] = df['NO2_ground'] / df['NO2_satellite'].replace(0, pd.NA)
df['SO2_ratio'] = df['SO2_ground'] / df['SO2_satellite'].replace(0, pd.NA)
df['CO_ratio'] = df['CO_ground'] / df['CO_satellite'].replace(0, pd.NA)
df['O3_ratio'] = df['O3_ground'] / df['O3_satellite'].replace(0, pd.NA)

# Drop extreme/infinite values
df = df.replace([float('inf'), -float('inf')], pd.NA).dropna()
```

```python

# Save cleaned dataset
output_path = "cleaned_aqi_merged_dataset.csv"
df.to_csv(output_path, index=False)
output_path
```
